```python
import cv2
import numpy as np
from keras.applications.vgg16 import VGG16,preprocess_input,decode_predictions
from keras.preprocessing import image
```

```python
model = VGG16(weights="imagenet") 
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()  # ret -> return, checks if the video feed is received, frame captures the images
    if not ret:
        break

    x = cv2.resize(frame, (224, 224))
    x = np.expand_dims(x, axis=0)  # Expand along the x-axis by adding one dimension as Keras expects a tensor
    x = preprocess_input(x)

    predictions = model.predict(x)
    label = decode_predictions(predictions, top=1)[0][0]  # Return the highest value, top=1
    label_name, label_confidence = label[1], label[2]

    cv2.putText(frame, f"{label_name}({label_confidence*100:.2f}%)", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1,
                (0, 255, 0), 2, cv2.LINE_AA)
    cv2.imshow("frame", frame)

    if cv2.waitKey(5) & 0xFF == ord("q"):  # Capture a frame every 5 ms, start the video if the 'q' key is pressed
        break

cap.release()  # Close the camera
cv2.destroyAllWindows()  # Close all open windows
```
